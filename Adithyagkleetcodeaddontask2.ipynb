```python
#Question 1. Two array Sum
#Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

##You may assume that each input would have exactly one solution, and you may not use the same element twice.

##You can return the answer in any order.
```

```python
def two_sum(a,b):
    length_number=len(a)
    for i in range(length_number):
        target=b-a[i]
        if target in a[i+1:]:
            return[i,a.index(target,i+1)]

def main():
    n=int(input("Enter the number of element:"))
    a = list(map(int, input("Enter elements separated by space: ").split()))
    b=int(input(("Enter the target number:")))
    result=two_sum(a,b)
    print("Result:",result)


if __name__ == "__main__":
    main()
```

```python
#question 3.
#Palindrome Number
#Given an integer x, return true if x is a palindrome, and false otherwise
```

```python
def palindrome(n):
    if n<0:
        return False
    orginal_number=n
    reverse_number=0
    while n>0:
        reverse_number=reverse_number*10+n%10
        n//=10
    return orginal_number==reverse_number
def main():
    num=int(input("Enter the number:"))
    if palindrome(num):
        print("True-It is palindrome number")
    else:
        print("False-It is not palindrome number")
if __name__ == "__main__":
    main()
    
```

```python
#Question4 .Roman to integer
```

```python
def Roman_integer(n):
    roman={
        'I':1,
        'II':2,
        'V':5,
        'IX':9,
        'X':10
    }
    total=0
    previous_value=0
    for i in reversed(n):
        value=roman[i]
        if value<previous_value:
            total-=value
        else:
            total+=value
        previous_value=value
    return total

print("The converting roman to intger for I is:",Roman_integer("I"))
print("The converting roman to intger for II is:",Roman_integer("II"))
print("The converting roman to intger for V is:",Roman_integer("V"))
print("The converting roman to intger for IX is:",Roman_integer("IX"))
print("The converting roman to intger for X is:",Roman_integer("X"))
```

```python
#Question5. Remove Duplicates from Sorted List
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head):
        current = head
        while current and current.next:
            if current.val == current.next.val:
                current.next = current.next.next
            else:
                current = current.next
        return head
def build_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for value in arr[1:]:
        current.next = ListNode(value)
        current = current.next
    return head
def to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result
def main():
    arr =list(map(int, input("Enter elements separated by space: ").split()))

    head = build_list(arr)
    solution = Solution()
    new_head = solution.deleteDuplicates(head)

    print("Output:", to_list(new_head))
if __name__ == "__main__":
    main()
```

```python
def remove_element(num,val):
    k=0
    for x in num:
        if x!=val:
            num[k]=x
            k+=1
    return k
def main():
    num=list(map(int, input("Enter elements separated by space: ").split()))
    val=int(input("Enter the element want to delete:"))
    k=remove_element(num,val)
    print("The value of k is:",k)
    print("The Required Array is:",num[:k])

if __name__ == "__main__":
    main()
```

```python
#Question7. Valid Palindrome using string
```

```python
def palindrome(n):
    string=""
    for ch in n:
        if ch.isalnum():
            string+=ch.lower()
    reverse_string=""
    for ch in string:
        reverse_string=ch+reverse_string
    return string==reverse_string
def main():
    a=input("Enter the string:")
    if palindrome(a):
        print("True-It is palindrome")
    else:
        print("False-It is not palindrome")

if __name__ == "__main__":
    main()
```

```python
#Question9.Add Digits
#Given an integer num, repeatedly add all its digits until the result has only one digit, and return it
```

```python
def add_digits(n):
    while n >= 10:
        total = 0
        temp = n   
        while temp > 0:
            total += temp % 10
            temp //= 10
        n = total 
    return n

def main():
    n = int(input("Enter number: "))
    print("The single digit is:", add_digits(n))

if __name__ == "__main__":
    main()
```

```python
def reverse_integer(n):
    sign=-1 if n<0 else 1
    n=abs(n)
    reverse=0
    while n!=0:
        reverse=reverse*10+n%10
        n//=10
    return sign*reverse
num=int(input("Enter the integer:"))
a=reverse_integer(num)
print("Reverse an integer is:",a)
```

```python
#Question 14.converting integer to roman
```

```python
def integer_to_roman(n):
    val=[1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    symbol=["M","CM","D","CD","C","XC","L","XL","X","IX","V","IV","I"]
    roman=""
    for i in range(len(val)):
        while n>=val[i]:
            n-=val[i]
            roman+=symbol[i]
    return roman
def main():
    n=int(input("Enter the number of range between 0 to 1050:"))
    a=integer_to_roman(n)
    print("After converting integer to roman:",a)
if __name__ == "__main__":
    main()
```

```python
def plus_one(n):
    index=[i for i in range(len(n))]
    for i in index[::-1]:
        if n[i]<9:
            n[i]+=1
            return n
    return 1+n
def main():
    n=a=list(map(int, input("Enter elements: ").split()))
    result=plus_one(n)
    print("The result is:",result)
if __name__ == "__main__":
    main()
    
```

```python
#Question 17. Sqrt(x)
#Given a non-negative integer x, return the square root of x rounded down to the nearest integer.
#The returned integer should be non-negative as well.
```

```python
def square_root(n):
    if n<2:
        return n
    left=1
    right=n//2
    while left<=right:
        mid=(left+right)//2
        a=mid*mid
        if a==n:
            return mid
        elif a<n:
            left=mid+1
        else:
            right=mid-1
    return right
def main():
    n=int(input("Enter the number:"))
    result=square_root(n)
    print("The square root of these number is:",result)

if __name__ == "__main__":
    main()           
```

```python
#Question18.. Longest Substring Without Repeating Characters
```

```python
def longest_substring(n):
    character=set()
    left=0
    maximum=0
    for i in range(len(n)):
        while n[i] in character:
            character.remove(n[left])
            left+=1
        character.add(n[i])
        maximum=max(maximum,i-left+1)
    return maximum
def main():
    n=input("Enter the string:")
    result=longest_substring(n)
    print("Longest substring is:",result)
if __name__ == "__main__":
    main()
    
```

```python
#Question19.climbing stairs
```

```python
def climbStairs(n):
    if n <= 2:
        return n
    a, b = 1, 2 
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b
def main():
    n = int(input("Enter number of stairs: "))
    result = climbStairs(n)
    print("Number of ways:", result)
if __name__ == "__main__":
    main()
```

```python
#Question 20.zigzag conversion
```

```python
def convert(s, numRows):
    if numRows == 1:
        return s
    rows = [""] * numRows    
    row = 0                 
    step = 1                
    for char in s:
        rows[row] += char 
        if row == 0:
            step = 1
        elif row == numRows - 1:
            step = -1
        row += step        
    return "".join(rows)   


def main():
    s = input("Enter string: ")
    numRows = int(input("Enter number of rows: "))
    result = convert(s, numRows)
    print("Zigzag:", result)


if __name__ == "__main__":
    main()
```
