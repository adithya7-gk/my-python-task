```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df=pd.read_csv(r"C:\Users\adith\Downloads\advertising.csv")
print(df.head())
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
z=LinearRegression()
z.fit(x,y)
y_predict=z.predict(x)
print("Intercept:",z.intercept_)
print("Coefficient:",z.coef_)
print("Mean Squared Error:", mean_squared_error(y, y_predict))
print("R^2 Score:", r2_score(y, y_predict))


```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df=pd.read_csv(r"C:\Users\adith\Downloads\HousingData.csv")
df=df.fillna(df.mean())
print(df.head())
x=df.drop('MEDV',axis=1)
y=df['MEDV']
z=LinearRegression()
z.fit(x,y)
y_predict=z.predict(x)
print("\nboston house dataset")
print("Intercept:",z.intercept_)
print("Coefficient:\n",z.coef_)
print("\nMean Squared Error:", mean_squared_error(y, y_predict))
print("\nR^2 Score:", r2_score(y, y_predict))
```

```python
#Wine Quality Test
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df=pd.read_csv(r"C:\Users\adith\Downloads\winequalityN.csv")
df['type'] = df['type'].map({'red': 0, 'white': 1})
df=df.fillna(df.mean())
print(df.head())
x=df.drop('quality',axis=1)
y=df['quality']
z=LinearRegression()
z.fit(x,y)
y_predict=z.predict(x)
print("\n Wine Quality Dataset")
print("Intercept:",z.intercept_)
print("\nCoefficient:\n",z.coef_)
print("\nMean Squared Error:", mean_squared_error(y, y_predict))
print("\nR^2 Score:", r2_score(y, y_predict))

```
