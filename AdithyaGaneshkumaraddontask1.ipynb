```python
#Questio 3. Fill the null values in respective columns like below:

    - apple(1kg): Mean value of that column
    - banana(1 dozen): Mean value of that column
    - grapes(1kg): Median value of that column
    - mango(1kg): Median value of that column
    - Water Melons(1): Fill null values with "Not Available".
    
Finally store the output to 'new_df' variable
```

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\adith\OneDrive\Desktop\fruits_data.csv")
new_df=df.copy()
new_df['apple(1kg)']=new_df['apple(1kg)'].fillna(new_df['apple(1kg)'].mean())
new_df['banana(1 dozen)']=new_df['banana(1 dozen)'].fillna(new_df['banana(1 dozen)'].mean())
new_df['grapes(1kg)']=new_df['grapes(1kg)'].fillna(new_df['grapes(1kg)'].median())
new_df['Water Melons(1)'] = new_df['Water Melons(1)'].fillna("Not Available")
print("After updating the data is:\n",new_df)
```
